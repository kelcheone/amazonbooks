```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

booksDf = pd.read_csv("Golang_books.csv")
booksDf['Price'].dtype
# migrating Rating from object to float
booksDf['Rating'] = pd.to_numeric(booksDf['Rating'], errors='coerce')
# getting all rows with null rating
booksDf[booksDf['Rating'].isnull()]
# dropping all rows with null rating
booksDf = booksDf.dropna()
# dropping all rows with ratings which are more than 5
booksDf = booksDf[booksDf['Rating'] <= 5]


# craete a bar graph for min maxes, mean
minp = booksDf['Price'].min()
maxp = booksDf['Price'].max()
meanp = booksDf['Price'].mean()
minr = booksDf['Rating'].min()
maxr = booksDf['Rating'].max()
meanr = booksDf['Rating'].mean()

colors = ['red', 'yellow', 'blue']
names = np.array(['min', 'max', 'mean'])
values = np.array([minp, maxp, meanp])
plt.bar(names, values, color=colors)
# save the plot as a png image
plt.savefig("Golang_books_bar.png")
plt.show()
rnames = np.array(['min', 'max', 'mean'])
rvalues = np.array([minr, maxr, meanr])
plt.bar(rnames, rvalues, color=colors)
# save the plot as a png image

plt.show()

# 
```
